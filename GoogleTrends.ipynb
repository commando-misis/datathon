```python
import pandas as pd
import pytrends
from pytrends.request import TrendReq
from pytrends import dailydata
```

```python
pytrend = TrendReq(timeout=(10,25), retries=5)
```

Использую метод get_daily_data, так как стандартный interest_over_time возвращает данные на конец каждой недели в место ежедневных

Выгрузку сделал с начала года, как пример, тк дальше ругается, что слишком много запросов, если за 3 года делать (в get_daily_data параметрами задаются начало и конец по месяцам и годам)

Структура данных, получаемых в результате каждого обращения к гуглу: https://stackoverflow.com/questions/62286165/pytrend-get-daily-data-column-meaning

```python
def get_google_stats(lst_key_words):
    """Список из слов для поиска в гугле преобразовывается в один датафрейм со значениями частоты запросов по каждому из них"""
    df_merged = pd.DataFrame()
    count = 0
    for kw in lst_key_words:
        res = pd.DataFrame()
        res = dailydata.get_daily_data(kw, start_year=2021, start_mon=1, stop_year=2021, stop_mon=10, geo = '')
        res.drop(labels=['isPartial', 'scale', kw+'_monthly', kw+'_unscaled'], axis=1, inplace=True)
        if count == 0:
            df_merged = res
            count += 1
            continue
        df_merged = df_merged.join(res)  
    return df_merged

keywords =['Bitcoin', 'Ethereum', 'Binance Coin', 'Cardano', 'Tether', 'XRP', 'Solana']
df_cur_stat = get_google_stats(keywords)
```
