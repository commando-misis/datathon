```python
import pandas as pd
```

```python
df = pd.read_csv('dataframe_final_merged.csv')
```

```python
normalized_df = df.set_index('date')
normalized_df = (normalized_df - normalized_df.mean()) / normalized_df.std()
```

```python
ax = normalized_df.plot(y=["XRP_price_USD", "XRP_google_trends"], label=['Цена XRP', 'Популярность XRP в Google'], figsize=(20, 10))
ax.set_xlabel("Дата")
ax.set_ylabel("Нормализованные значения цены и популярности")
ax.set_title("Цена Ripple (XRP) и популярность в поисковых системах")
```

```python
fig = ax.get_figure()
fig.savefig('plots/xrp_price_google_trends.jpg')
```

```python
ax = normalized_df.plot(y=["XRP_price_USD", "XRP_reddit_mentions"], label=['Цена XRP', 'Упоминания XRP в комментариях Reddit'], figsize=(20, 10))
ax.set_xlabel("Дата")
ax.set_ylabel("Нормализованные значения цены и упоминаний")
ax.set_title("Цена Ripple (XRP) и упоминания в комментариях Reddit")
```

```python
fig = ax.get_figure()
fig.savefig('plots/xrp_price_reddit.jpg')
```
