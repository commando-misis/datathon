```python
import pandas as pd
import numpy as np
```

```python
# Read values as str so pandas doesn't cut off float precision
final_df = pd.read_csv('final_2020-01-01_2021-09-30.csv', dtype='str')
```

```python
trends = pd.read_csv('daily_search_volume_2019-12_2021-09.csv', float_precision='round_trip')
```

```python
# Rename columns to be consistent with our naming scheme
trends = trends.set_axis(['date', 'BTC_google_trends', 'ETH_google_trends', 'BNB_google_trends', 'ADA_google_trends', 'USDT_google_trends', 'XRP_google_trends', 'SOL_google_trends'], axis=1, inplace=False)
```

```python
# Merge dataframes
merged_df = final_df.merge(trends)
```

```python
# Drop SOL google trends for dates it didn't exist
# Warning - hardcoded range!
merged_df.iloc[0:101].loc[:,'SOL_google_trends'] = np.nan
```

```python
cols = merged_df.columns.tolist()

# Rearange columns
cols = cols[0:1] + cols[1:11] + [cols[-2]] \
+ cols[11:21] + [cols[-1]] \
+ cols[21:31] + [cols[-5]] \
+ cols[31:41] + [cols[-4]] \
+ cols[41:51] + [cols[-6]] \
+ cols[51:61] + [cols[-7]] \
+ cols[61:71] + [cols[-3]] 
```

```python
merged_df = merged_df[cols]
```
